### customer_id: key to the customer dataset. Each order has a unique customer_id.

```python
# assign the orders key in the dict to a orders_df variable

orders_df = data['orders']
orders_df.head()
```

```python
# Check the data type of each column in orders_df

orders_df.info()
```

```python
# change to datetime format

orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'], errors='coerce')
orders_df['order_approved_at'] = pd.to_datetime(orders_df['order_approved_at'], errors='coerce')
orders_df['order_delivered_carrier_date'] = pd.to_datetime(orders_df['order_delivered_carrier_date'], errors='coerce')
orders_df['order_delivered_customer_date'] = pd.to_datetime(orders_df['order_delivered_customer_date'], errors='coerce')
orders_df['order_estimated_delivery_date'] = pd.to_datetime(orders_df['order_estimated_delivery_date'], errors='coerce')
```

```python
# validate the datetime change

orders_df.info()
```

```python
# Check the null values

orders_df.isna().sum()
```

```python
# creating the dataframe for the null values in the orders_df

orders_mask = (orders_df['order_delivered_carrier_date'].isna() |
              orders_df['order_approved_at'].isna() |
              orders_df['order_delivered_customer_date'].isna())


orders_df.loc[orders_mask].head(500)
```

```python
# Checking if there is a 'delivered' status in the null dataframe for orders_df

delivered_mask = (
    (orders_df['order_status'] == 'delivered') &
    (orders_df['order_delivered_carrier_date'].isna()) &
    (orders_df['order_approved_at'].isna()) &
    (orders_df['order_delivered_customer_date'].isna())
)

orders_df.loc[delivered_mask]
```

```python
# calculating the sum of na vlaues in the orders_df for validation

carrier_sum = orders_df['order_delivered_carrier_date'].isna().sum()
ordered_sum = orders_df['order_approved_at'].isna().sum()
customer_sum = orders_df['order_delivered_customer_date'].isna().sum()

orders_na_sum = carrier_sum + ordered_sum + customer_sum
orders_na_sum
```

```python
# validate orders_df dataframe

print(f'orders dataframe has {orders_df.shape[0]} rows and {orders_df.shape[1]} columns')
print(f'orders dataframe has {orders_df.isna().sum().sum()} null values')
print(f'orders dataframe has {orders_df.duplicated().sum()} duplicated rows')
```
